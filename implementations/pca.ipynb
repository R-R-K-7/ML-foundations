```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class PCA():
    def __init__(self,n_components):
        self.components = None
        self.n_components = n_components
        self.mean = self.std = None
    def fit(self,X):
        self.mean = np.mean(X,axis=0)
        self.std = np.std(X,axis=0)
        self.std[self.std==0] = 1.0
        X_centered = (X-self.mean)/self.std
        n = X_centered.shape[0]
        covariance_mat = np.dot(X_centered.T,X_centered)/(n-1)
        eigenval,eigenvec = np.linalg.eig(covariance_mat)
        ind = np.argsort(eigenval)[::-1]
        eigenval = eigenval[ind]
        eigenvec = eigenvec[:,ind]
        self.components = eigenvec[:,:self.n_components]
        self.full_variance_ratio = eigenval/np.sum(eigenval)
    def transform(self,X):
        X_centered = (X-self.mean)/self.std
        return np.dot(X_centered,self.components)
```

# 1. Create Dummy Data
# A random diagonal blob
np.random.seed(42)
x1 = np.random.rand(100) * 10
x2 = x1 + np.random.normal(0, 2, 100) # x2 is correlated with x1
X = np.column_stack((x1, x2))

# 2. Run Our PCA
# We want to reduce 2 dimensions down to 1 (just the main line)
pca = PCA(n_components=1)
pca.fit(X)
X_projected = pca.transform(X)

print(f"Original Shape: {X.shape}")
print(f"Projected Shape: {X_projected.shape}")
print(f"Explained Variance Ratio: {pca.full_variance_ratio}")

# 3. Visualization
plt.figure(figsize=(10, 4))

# Plot Original Data
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], color='blue', alpha=0.5)
plt.title("Original Data (2D)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

# Plot Projected Data (1D representation)
# We set Y-axis to 0 strictly to visualize the 1D line
plt.subplot(1, 2, 2)
plt.scatter(X_projected, np.zeros_like(X_projected), color='red', alpha=0.5)
plt.title("Projected Data (1D)")
plt.xlabel("Principal Component 1")
plt.yticks([])

plt.tight_layout()
plt.show()
